## Plot variable trajectories for a given person

```python
id = 0
subject = df[df['id'] == id]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

vbls = ['HR', 'O2Sat', 'SBP']
for i, v in enumerate(vbls):
    ax[i].plot(subject[v])
    ax[i].set_title(v, fontsize=18, fontweight='bold')
```

```python
vbls = ['HR', 'O2Sat', 'Temp', 'SBP', 'MAP', 'DBP', 'Resp']
n_cols = 4
n_rows = 2

fig, ax = plt.subplots(n_rows, n_cols, figsize=(30, 10))
ax = ax.ravel()

for i, vbl in enumerate(vbls):
    ax[i].plot(subject[vbl], label='raw')
    ax[i].legend()
    ax[i].set_title(vbl, fontsize=18)
```
